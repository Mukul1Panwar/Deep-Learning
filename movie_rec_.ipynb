```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/netflix_titles.csv")
```

```python
df.head()
```

```python

df.insert(0, 'index', range(len(df)))
```

```python
df.shape
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
features = ['title','listed_in','description']
```

```python
for f in features:
  df[f] = df[f].fillna('')
```

```python
combined_features = df['title']+' '+df['listed_in']+' '+df['description']
```

```python
combined_features
```

```python
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vec = TfidfVectorizer()
```

```python
feature_vec = vec.fit_transform(combined_features)
```

```python
print(feature_vec)
```

```python
feature_vec.shape
```

```python
similarity = cosine_similarity(feature_vec)
```

```python
similarity
```

```python
movie_name = input("Enter the movie name : ")
```

```python
title_list = df['title'].tolist()
```

```python
print(title_list)
```

```python
find_close_match = difflib.get_close_matches(movie_name,title_list)
```

```python
print(find_close_match)
```

```python
close_match = find_close_match[0]
```

```python
index_of_movie = df[df.title == close_match]['index'].values[0]
```

```python
print(index_of_movie)
```

```python
similiraty_score = list(enumerate(similarity[index_of_movie]))
print(similiraty_score)
```

```python
len(similarity)
```

```python
sorted_similiraty = sorted(similiraty_score,key = lambda x:x[1],reverse = True)
```

```python
print(sorted_similiraty)
```

```python
i = 1
for movie in sorted_similiraty:
  index = movie[0]
  title_from_index = df[df.index==index]['title'].values[0]
  if (i<31):
    print(i, ".", title_from_index)
    i+=1
```

```python
movie_name = input("enter the movie name : ")
movies_title = df['title'].tolist()
close_movie = difflib.get_close_matches(movie_name,movies_title)
close_movie = close_movie[0]
movie_index = df[df.title == close_movie]['index'].values[0]
similiraty_score = list(enumerate(similarity[movie_index]))
sorted_similarity_score = sorted(similiraty_score,key = lambda x: x[1],reverse=True)

i = 1
for movie in sorted_similarity_score:
  index = movie[0]
  title_to_index = df[df.index == index]['title'].values[0]
  if (i<11):
    print(i,".", title_to_index)
    i+=1




```
